This notebook is intended to learn the basics of opencv:
    
- opening a video
- getting a frame from the video
- drawing a box on the frame and show it 

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import json
```

```python
cv.__version__
```

Create a video capture object and read a frame from it

```python
video_file = "../../data/input.mkv"
video_capture = cv.VideoCapture(video_file)
```

```python
ret, frame = video_capture.read()
ret
```

Convert the image color palette from BGR to RGB and show it

```python
frame = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
plt.imshow(frame)
```

Read the json file with the bounding boxes 

```python
json_file = "../../data/initial_conditions.json"
with open(json_file) as f:
  initial_conditions = json.load(f)
```

```python
x, y, w, h = initial_conditions[0]["coordinates"]
```

Draw the rectangles on the first frame and show it

```python
for obj in initial_conditions:
    x, y, w, h = obj["coordinates"]
    cv.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 10)
```

```python
plt.imshow(frame)
```

Read some properties of the video

```python
frame_count = video_capture.get(cv.CAP_PROP_FRAME_COUNT)
frame_fps = video_capture.get(cv.CAP_PROP_FPS)
frame_width = video_capture.get(cv.CAP_PROP_FRAME_WIDTH)
frame_height = video_capture.get(cv.CAP_PROP_FRAME_HEIGHT)
frame_format = video_capture.get(cv.CAP_PROP_FORMAT)
```

```python
print(frame_count)
print(frame_fps)
print(frame_height)
print(frame_width)
```

Release the video capture

```python
video_capture.release()
```
